# 01. 서비스 이탈예측 데이터

```python
import pandas as pd
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_test.csv")
```

# 01. EDA

```python
# 데이터 샘플 확인
display(x_train.head().T)
display(y_train.head().T)
```

```python
# 데이터 타입 & 기초 통계
display(x_train.info())
display(x_train.describe())
```

```python
# 결측치
display(x_train.isnull().sum())
display(x_test.isnull().sum())
```

# 02. 데이터 전처리

```python
# 결측치 처리
```

```python
# 이상치 처리
```

```python
# Drop - ID
drop_col = ['CustomerId','Surname']
x_train_drop = x_train.drop(columns = drop_col)
x_test_drop = x_test.drop(columns = drop_col)
y = y_train['Exited']
```

# 03. 피처 엔지니어링

```python
from sklearn.model_selection import train_test_split
```

```python
# 범주형 변수 - Encoding

x_train_dum = pd.get_dummies(x_train_drop)
x_test_dum = pd.get_dummies(x_test_drop)
x_test_dum = x_test_dum[x_train_dum.columns] # train-test col순서 정렬
```

```python
# 수치형 변수 - Scaling
```

# 04. Modeling

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score , f1_score, recall_score, roc_auc_score ,precision_score
```

```python
# 데이터 분리 : 학습용 - 검증용
X_train, X_validation, Y_train, Y_validation = train_test_split(x_train_dum, y, test_size=0.3, random_state=42)
```

```python
# 학습
rf = RandomForestClassifier(random_state =23)
rf.fit(X_train,Y_train)
```

```python
# 검증 - accuracy , f1_score , recall , precision -> model.predict로 결과뽑기

predict_validation_label = rf.predict(X_validation)

print(accuracy_score(Y_validation,predict_validation_label))
print(f1_score(Y_validation,predict_validation_label))
print(recall_score(Y_validation,predict_validation_label))
print(precision_score(Y_validation,predict_validation_label))
```

```python
# 검증 - auc , 확률 -> model.predict_proba()[:,1]

predict_validation_prob = rf.predict_proba(X_validation)[:,1]
print(roc_auc_score(Y_validation,predict_validation_prob))
```

```python
# 예측

pred = rf.predict(x_test_dum)
output = pd.DataFrame({'idx': x_test_dum.index, 'Outcome': pred})
output.head()
```

```python
# 제출

output.to_csv('1234567.csv', index=False)
```
